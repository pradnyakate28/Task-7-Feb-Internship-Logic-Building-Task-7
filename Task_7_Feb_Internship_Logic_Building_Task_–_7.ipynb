```python
# Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # default status

    def switch(self, action):
        if action.upper() == "ON":
            self.status = "ON"
        else:
            self.status = "OFF"

    def show_status(self):
        print(self.name + " is " + self.status)

# Taking input
light_name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(light_name)
light.switch(action)
light.show_status()
```

```python
# Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display()
```

```python
# Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

# Taking input
name = input("Contact Name: ")
phone = input("Phone Number: ")

c1 = Contact(name, phone)
c1.show_contact()
```

```python
# Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def generate_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

# Taking input
product_name = input("Product Name: ")
price = int(input("Price: "))

item = Product(product_name, price)
item.generate_tag()
```

```python
# Problem 5: Movie Rating Display System

class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")

# Taking input
movie_name = input("Movie Name: ")
rating = float(input("Rating: "))

m1 = Movie(movie_name, rating)
m1.display_movie()
```

```python
# Problem 6: Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

# Taking input
customer = input("Customer Name: ")
address = input("Address: ")

d1 = Delivery(customer, address)
d1.print_delivery()
```
